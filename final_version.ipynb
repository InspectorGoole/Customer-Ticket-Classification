```python
from transformers import AutoModel, AutoTokenizer
from transformers import AutoModelForSequenceClassification
```

```python
model = AutoModelForSequenceClassification.from_pretrained('distilbert-base-uncased', num_labels =5)
tokenizer = AutoTokenizer.from_pretrained('distilbert-base-uncased')
```

```python
from sklearn.model_selection import train_test_split

train_data, test_data = train_test_split(data, train_size=0.8, random_state=0)
```

```python
print("Customer query:")
print(test_data['body'].iloc[12])
print("\nTicket Type:")
print(test_data['queue'].iloc[12])
```

```python
from sklearn.preprocessing import LabelEncoder
import torch

label_encoder = LabelEncoder()

train_labels = label_encoder.fit_transform(train_data['queue'])
test_labels = label_encoder.transform(test_data['queue'])

tokenized_training_data = tokenizer(train_data['body'].tolist(), return_tensors='pt', padding= True, truncation = True, max_length=64)
tokenized_testing_data = tokenizer(test_data['body'].tolist(), return_tensors='pt', padding=True, truncation=True, max_length=64)

tokenized_training_data['labels'] = torch.tensor(train_labels, dtype=torch.long)
tokenized_testing_data['labels'] = torch.tensor(test_labels, dtype=torch.long)
```

```python
label_mapping = dict(zip(label_encoder.classes_, label_encoder.transform(label_encoder.classes_)))
print("Label Encoding Mapping:")
for label, encoded_value in label_mapping.items():
    print(f"{label}: {encoded_value}")
```

```python
from transformers import Trainer, TrainingArguments

training_args = TrainingArguments(
    output_dir="/content/sample_data",
    eval_strategy="epoch",
    num_train_epochs=5,
    learning_rate=3e-5,
    per_device_train_batch_size=16,
    per_device_eval_batch_size=16,
    weight_decay=0.02,
    warmup_steps = 500,
    save_strategy = "epoch",
    load_best_model_at_end = True,
    metric_for_best_model= "f1",
    report_to='none' # Disable Weights & Biases logging
)
```

```python
from datasets import Dataset

train_dataset = Dataset.from_dict(tokenized_training_data)
test_dataset = Dataset.from_dict(tokenized_testing_data)
```

```python
from sklearn.metrics import accuracy_score, precision_recall_fscore_support, classification_report
import numpy as np

# Define a function to compute metrics
def compute_metrics(eval_pred):
    logits, labels = eval_pred
    predictions = np.argmax(logits, axis=-1)

    precision, recall, f1, _ = precision_recall_fscore_support(labels, predictions, average='weighted')
    acc = accuracy_score(labels, predictions)

    return {
        'accuracy': acc,
        'f1': f1,
        'precision': precision,
        'recall': recall
    }

trainer = Trainer(
    model = model,
    args = training_args,
    train_dataset= train_dataset,
    eval_dataset = test_dataset,
    compute_metrics=compute_metrics,
    tokenizer= tokenizer
)

trainer.train()
```

```python

results = trainer.evaluate()
print(results)

# For detailed per-class metrics
predictions = trainer.predict(test_dataset)
y_pred = np.argmax(predictions.predictions, axis=-1)
y_true = predictions.label_ids

# Get category names

labels = ['Billing and Payments', 'Customer Service', 'Returns and Exchanges', 'Service Outages and Maintenance', 'Technical Support']

# Detailed report
print(classification_report(y_true, y_pred, target_names=labels))
```

```python
new_input = tokenizer(new_data, return_tensors='pt', padding=True, truncation=True, max_length=64)

# Move the input tensors to the same device as the model
device = torch.device("cuda" if torch.cuda.is_available() else "cpu")
new_input = {k: v.to(device) for k, v in new_input.items()}
model.to(device)

# new_input.items(): Iterates through key-value pairs in your input dictionary

# v.to(device): Moves each tensor to the GPU/CPU

# Result: All input tensors are now on the same device as your model
# Moves the entire model (all layers, parameters) to the specified device

with torch.no_grad():
  outputs = model(** new_input)

predicted_labels = torch.argmax(outputs.logits, dim=1).tolist()

label_map = { 0: 'Billing and Payments',
              1 : 'Customer Service',
              2 : 'Returns and Exchanges',
              3 : 'Service Outages and Maintenance',
              4 : 'Technical Support'}

for i, predicted_label in enumerate(predicted_labels):
  ticket_type = label_map[predicted_label]
  print(f"\n Input text {i+1}: {new_data[i]}" )
  print(f"\n Predicted label, {ticket_type}")
```
